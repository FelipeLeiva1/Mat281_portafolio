```python
age_f=df["age"].unique()
frec=pd.Series()
#for i in age_f:
for i in age_f:
    k=0
    for j in df["age"]:
        if i==j:
            k=k+1
    frec.loc[i]=k
f_min= frec.min()

mask = frec == f_min

frec[mask]

print("las edades con menos frecuencia son:")
for i in frec[mask].index:
    print(i)
```
